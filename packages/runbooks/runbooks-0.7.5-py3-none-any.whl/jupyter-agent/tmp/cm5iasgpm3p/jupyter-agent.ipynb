```python
import numpy as np

# Define the coefficients of the line
m = 2  # slope
c = 3  # intercept

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values
y = m * x + c

print("The equation of the line is: y = {:.2f}x + {:.2f}".format(m, c))

# Plot the line
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Equation y = {:.2f}x + {:.2f}'.format(m, c))
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.show()
```

This script first defines the coefficients of the line (slope and intercept), then generates x values over a range from -10 to 10. It calculates the corresponding y values using the equation of the line, prints out the equation of the line, and finally plots the line using matplotlib. The resulting plot shows the linear relationship between x and y according to the given equation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probabilities
P_A = 0.01  # Probability of having the disease
P_B_given_A = 0.99  # Probability of testing positive given that you have the disease
P_B_given_not_A = 0.02  # Probability of testing positive given that you don't have the disease

# Calculate P(not A)
P_not_A = 1 - P_A

# Calculate P(B) using the law of total probability
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A

# Apply Bayes' theorem to find P(A|B)
P_A_given_B = (P_B_given_A * P_A) / P_B

print("The probability of having the disease given a positive test result is: {:.2%}".format(P_A_given_B))

# Create an array of different prior probabilities
prior_probabilities = np.linspace(0, 1, 100)

# Calculate the posterior probabilities for each prior probability
posterior_probabilities = (P_B_given_A * prior_probabilities) / (P_B_given_A * prior_probabilities + P_B_given_not_A * (1 - prior_probabilities))

# Plot the results
plt.plot(prior_probabilities, posterior_probabilities)
plt.xlabel('Prior Probability of Having the Disease')
plt.ylabel('Posterior Probability of Having the Disease Given a Positive Test Result')
plt.title('Bayes\' Theorem Applied to Medical Testing')
plt.grid(True)
plt.show()
```

This script first defines the probabilities involved in the problem, then applies Bayes' theorem to calculate the probability of having the disease given a positive test result. It also creates an array of different prior probabilities and calculates the corresponding posterior probabilities using Bayes' theorem. Finally, it plots the relationship between the prior and posterior probabilities. The resulting plot shows how the posterior probability changes as the prior probability varies.
