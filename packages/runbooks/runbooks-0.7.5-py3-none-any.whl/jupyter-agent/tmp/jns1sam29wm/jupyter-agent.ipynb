```python
import numpy as np
import matplotlib.pyplot as plt

# Define the prior probability distribution
def prior(theta):
    return np.ones(len(theta)) / len(theta)

# Define the likelihood function
def likelihood(theta, x):
    return np.exp(-((theta - x) ** 2) / (2 * 0.1))

# Define the posterior probability distribution using Bayes' theorem
def posterior(theta, x):
    prior_prob = prior(theta)
    likelihood_prob = likelihood(theta, x)
    evidence = np.sum(prior_prob * likelihood_prob)
    return likelihood_prob * prior_prob / evidence

# Generate a range of theta values
theta = np.linspace(0, 1, 100)

# Calculate the posterior probabilities for different x values
posterior_probs = []
for x in [0.2, 0.5, 0.8]:
    posterior_prob = posterior(theta, x)
    posterior_probs.append(posterior_prob)

# Plot the posterior probability distributions
plt.plot(theta, posterior_probs[0], label='x=0.2')
plt.plot(theta, posterior_probs[1], label='x=0.5')
plt.plot(theta, posterior_probs[2], label='x=0.8')
plt.xlabel('Theta')
plt.ylabel('Posterior Probability')
plt.title('Posterior Probability Distributions')
plt.legend()
plt.show()
```

This code first defines the prior probability distribution, likelihood function, and posterior probability distribution according to Bayes' theorem. It then generates a range of theta values and calculates the posterior probabilities for different x values. Finally, it plots the posterior probability distributions for these x values.

The resulting plot shows how the posterior probability distribution changes as the observed value x changes. The peak of the posterior distribution represents the most likely value of theta given the observation x. This demonstrates how Bayes' theorem updates our belief about the parameter theta based on new observations.
