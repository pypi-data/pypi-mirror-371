# 一、两独立样本均值之差检验

```python
import abetter as ab

# 定义均值型指标样本（其中样本量n、样本均值mean、样本标准差std，是3个必要参数，其他为可选参数）
s1 = ab.SampleMean(n=1000000, mean=0.5000, std=0.2000, field_name='订单量', group_name='实验组', group_ratio=0.1, report_title="AB实验结果")
s2 = ab.SampleMean(n=2000000, mean=0.5005, std=0.2010, field_name='订单量', group_name='空白组', group_ratio=0.2)

# 检验
ss = s1 - s2
ss   # Pandas数据框存储于 ss.data_all
```

```python
s1 = ab.SampleMean(1000, 0.50, 0.20, field_name='订单量', group_name='实验组', group_ratio=0.1, report_title="AB实验结果")
s2 = ab.SampleMean(2000, 0.55, 0.21, field_name='订单量', group_name='空白组', group_ratio=0.2)
ss = s1 - s2
ss
```

```python
s1 = ab.SampleMean(10, 0.50, 0.20, field_name='订单量', group_name='实验组', group_ratio=0.1, report_title="AB实验结果")
s2 = ab.SampleMean(20, 0.55, 0.21, field_name='订单量', group_name='空白组', group_ratio=0.2)
ss = s1 - s2
ss
```

```python
s1 = ab.SampleMean(10, 0.50, 0.20, field_name='订单量', group_name='实验组', report_title="AB实验结果")
s2 = ab.SampleMean(20, 0.55, 0.21, field_name='订单量', group_name='空白组')
ss = s1 - s2
ss
```

# 二、两独立样本比率之差检验

## 2.1 两独立样本比率之差检验（大样本）
每组正负样本量都>=10

```python
import abetter as ab

# 定义比率型指标样本（其中样本量n、阳性样本量k，是2个必要参数，其他为可选参数）
s1 = ab.SampleProp(n=10000, k=60, field_name='交易人数', group_name='实验组', group_ratio=0.1, report_title="AB实验结果")
s2 = ab.SampleProp(n=20000, k=50, field_name='交易人数', group_name='空白组', group_ratio=0.2)

# 检验
ss = s1 - s2
ss   # Pandas数据框存储于 ss.data_all
```

```python
ss.data_best
```

```python
ss.data_all
```

```python
ss.data_flat
```

## 2.2 两独立样本比率之差检验（小样本）

```python
s1 = ab.SampleProp(100, 1, field_name='订单量', group_name='实验组', group_ratio=0.1, report_title="AB实验结果")
s2 = ab.SampleProp(200, 6, field_name='订单量', group_name='空白组', group_ratio=0.2)
ss = s1 - s2
ss
```
